### Fill in missing values

```python
for column in df_n:
    btao = df_n[column].isna().sum()
    btao2 = (btao * 100)/len(df_n)
    if btao2 > 0:
        df_n[column].fillna(float(df_n[column].mean()), inplace=True)
```
