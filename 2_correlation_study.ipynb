# Correlation study

Our dataset only contains numerical features since the number of numerical features alone is 227. Of a total of 248 features that we could have used, 21 were categorical with at least 2 categories. Encoding these would lead to an addition of at least 42 features (obviously more than 42 since only a few were binary categorical features). We already have 227 features to work with so we don't have to add more features since it will only increase number of computations i.e, curse of dimensionality.

The objective of doing a *feature correlation study* is to drop features that are highly correlated with other numerical featues. The threshold we will use for this is 0.9 (because correlation coefficient lies between 1 and -1 for Pearson correlation and 0.9 implies the features are highly positively correlated => as one feature grows the other grows at roughly the same rate, 90% )

Another point to note is that we need to do correlation study ON train data and not use the target variable. The reason being: we cannot do the same correlation study on the test dataset, so we must rely on the correlation between each pair of features.

## 1. Split data into x and y

```python
# we cannot do correlation on the target variable
# drop 'isFraud'
# this will create the x-y split required for modelling
# we then perform Pearson correlation on the x_vector

x = df.drop(columns=['isFraud'], axis=1)
y = df['isFraud']
```

```python
# feature correlation on x vector

# cor_matrix = x.corr(method='pearson')
# print(cor_matrix)
```

```python
# we cannot use a heatmap because of the large number of features.
# Our display will be unable to showcase all the correlation coefficients as we can see from the bove table

# we will drop features that are highly positively correlated (more than 0.9)
# we will keep negatively correlated features 
```

```python
# we need to filter out the highly correlated features and store them in a list so we can drop them from our dataset
# d -> dataset
# t -> threshold
# colcor -> set() because we need to avoid adding the same column name again 
# corr_mat -> it is the same correlation matrix as seen above
# we use j in range(i) because the iterations will run through each feature until the feature in i 
# this way each feature eventually gets compared wrt correlation with every other feature
# and, using a set ensures we don't get the same feature added again
# cn -> column name
# oth -> feature with which cn[i] has high correlation with
# r -> value of correlation coeff

def corl_func(d,t) :
    colcor = set()
    corr_mat = d.corr(method='pearson')
    for i in range(len(corr_mat.columns)) :
        for j in range(i) :
            if (corr_mat.iloc[i,j]) > t :
                cn = corr_mat.columns[i]
                colcor.add(cn)                
    return colcor
```

```python
# highly correlated features

high_corr_feat = corl_func(x,0.9)

len(high_corr_feat)
```

```python
# drop these 90 features from our train set
# drop these features directly from the original dataframe

df_train = df.drop(high_corr_feat, axis=1)
```

```python
# save this new dataframe (post correlation study) as a new csv file for modelling

df_train.to_csv("trainf.csv",index=False)
```
